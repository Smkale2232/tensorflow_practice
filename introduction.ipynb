### Tensor Shapes

```python
import tensorflow as tf
print(tf.version)
```

```python
t = tf.zeros([5, 5, 5, 5])
print(t)
```

```python
t = tf.reshape(t, [625])
print(t)
```

```python
t = tf.reshape(t, [5, -1])
print(t)
```

```python
t = tf.reshape(t, [125, 5])
print(t)
```

```python
t = tf.reshape(t, [25, -1])
print(t)
```
